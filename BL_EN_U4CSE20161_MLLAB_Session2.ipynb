```python
import nltk
from nltk.tokenize import word_tokenize
import pandas as pd
import numpy as np
nltk.download('punkt')
from nltk.corpus import stopwords 
nltk.download('stopwords')
```

```python
frame = pd.read_excel(r'/content/drive/MyDrive/5th SEMESTER INFORMATION/19CSE305 MACHINE LEARNING/AssignmentHW2/Lab Session2 Data.xlsx')
print(frame)
```

```python
token_set = frame.apply( lambda row : nltk.word_tokenize(row['Report']),axis=1)
print (token_set)
```

```python
main_set = list(token_set)
print(main_set)
```

```python
t_amount = set()
last_string=""
for i in range (0,37):
  length = len(main_set[i])
  for j in range(0,length):
    last_string = last_string + ' '+ main_set[i][j]
last_string = last_string.strip()
print(last_string)
```

```python
end_list = last_string.split(" ")
print(end_list)
print(len(end_list))
```

```python
t_amount = set(end_list)
print(t_amount)
print(len(t_amount))
```

```python
stopwords = nltk.corpus.stopwords.words('english')
stopwords.append('newWord')
for e in  stopwords:
  for aim_point in end_list:
    if(e == aim_point):
      end_list.remove(e)
print(end_list)
print(len(end_list))
```

```python
end_bag_of_words = set(end_list)
print(end_bag_of_words)
```

```python
print(len(end_bag_of_words))
```

```python
v1new=[]
for i in range(37):
    v1new.append([])
for ind in range(0,37):
  b=[]
  liststudent = (frame['Report'].values[ind]).split(" ")
  for e in set(liststudent):
    if e in stopwords:
      b.append(0)
    else:
      b.append(1)
  v1new[ind].extend(b)
print(v1new)
```

```python
v1 = []
for e in token_set:
  temp = []
  for word in list(end_bag_of_words):
    if(word in e):
      temp.append(1)
    else:
      temp.append(0)
  v1.append(temp)
for i in v1:
  print (i)
```

```python
v2=[]
for i in range(37):
	v2.append([])
 
for ind in range(0,37):
  b=[]
  liststudent = (frame['Report'].values[ind]).split(" ")
  for m in end_bag_of_words:
    b.append(liststudent.count(m))
  v2[ind].extend(b)
for i in v2:
  print(i)
```

```python
def jaccard_similarity(A, B):
    nominator = A.intersection(B)
    denominator = A.union(B)
    similarity = len(nominator)/len(denominator)
    return similarity
```

```python
result = 0
jaccard=[]
for i in v1new:
  b=[]
  for j in v2:
    result = jaccard_similarity(set(i),set(j))
    b.append(result)
  jaccard.append(b)
for i in jaccard:
  print(i)
```

```python
from numpy.linalg import norm
cosine_list=[]
for i in v1:
  b=[]
  for j in v2:
    cosine = np.dot(np.array(i),np.array(j))/(norm(np.array(i))*norm(np.array(j)))
    b.append(cosine)
  cosine_list.append(b)
for i in cosine_list:
  print(i)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(21,8))
fig.suptitle('Visual representation of similarity measues')
ax1.title.set_text('Jaccardian Similarity')
ax2.title.set_text('Cosine Similarity')
sns.heatmap(jaccard, ax=ax1, xticklabels=1, cmap=sns.light_palette("pink", as_cmap=True) , vmin=0, vmax=1)
sns.heatmap(cosine_list, ax=ax2, xticklabels=1, cmap=sns.light_palette("pink", as_cmap=True), vmin=0, vmax=1)
plt.show()
```
